```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Conv2D, Input

train_df = pd.read_csv("C:\\Users\\Dharani Ravi\\Downloads\\digit-recognizer\\train.csv")

x_train = train_df.iloc[:, 1:].values 
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0 


input_layer = Input(shape=(28, 28, 1))
conv_layer = Conv2D(filters=1, kernel_size=(3,3), activation=None)(input_layer)  
model = Model(inputs=input_layer, outputs=conv_layer)

feature_map = model.predict(np.expand_dims(x_train[0], axis=0))

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(x_train[0].squeeze(), cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(feature_map[0, :, :, 0], cmap='gray')
axes[1].set_title('Feature Map after Convolution')
axes[1].axis('off')

plt.show()
```
