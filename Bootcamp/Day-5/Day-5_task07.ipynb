```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances
import matplotlib.pyplot as plt

file_path = "C:\\Users\\Dharani Ravi\\Downloads\\kmeans\\Mall_Customers.csv"
df = pd.read_csv(file_path)

X = df[['Annual Income (k$)', 'Spending Score (1-100)']].values

metrics = ['euclidean', 'manhattan', 'cosine']
results = {}

for metric in metrics:
    kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
    if metric != 'euclidean':
        dist_matrix = pairwise_distances(X, metric=metric)
        clusters = kmeans.fit_predict(dist_matrix)
    else:
        clusters = kmeans.fit_predict(X)
    results[metric] = clusters

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for i, metric in enumerate(metrics):
    axes[i].scatter(X[:, 0], X[:, 1], c=results[metric], cmap='viridis', edgecolors='k', s=100)
    axes[i].set_title(f'K-means with {metric} distance')
    axes[i].set_xlabel('Annual Income (k$)')
    axes[i].set_ylabel('Spending Score (1-100)')
plt.show()

```
