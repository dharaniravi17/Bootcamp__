```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

file_path = "C:\\Users\\Dharani Ravi\\Downloads\\kmeans\\Mall_Customers.csv"
df = pd.read_csv(file_path)

X = df[['Annual Income (k$)', 'Spending Score (1-100)']].values

initializations = ['random', 'k-means++']
plt.figure(figsize=(12, 5))

for i, init in enumerate(initializations, 1):
    kmeans = KMeans(n_clusters=5, init=init, random_state=42, n_init=10)
    clusters = kmeans.fit_predict(X)
    
    plt.subplot(1, 2, i)
    plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o', edgecolors='k', s=100)
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
    plt.xlabel('Annual Income (k$)')
    plt.ylabel('Spending Score (1-100)')
    plt.title(f'K-means with {init} initialization')
    plt.legend()

plt.tight_layout()
plt.show()
```
